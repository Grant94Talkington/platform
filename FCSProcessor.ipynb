```python
import os
import FlowCytometryTools
from FlowCytometryTools import FCMeasurement
from pylab import *

datadir = r'/Users/granttalkington/Desktop/Coding_shit/Python_Projects/FCSfilereader'

datafile = r'/Users/granttalkington/Desktop/Coding_shit/Python_Projects/FCSfilereader/unstained.fcs'

sample = FCMeasurement(ID='unstained', datafile=datafile)

ChannelNames = ['SSC-H', 'SSC-A', 'FSC-H', 'FSC-A', 'SSC-B-H', 'SSC-B-A', 'Pacific Blue-A', 'BV510-A', 'BV711-A', 'BV785-A', 'PE-Cy7-A', 'Alexa Fluor 700-A', 'Zombie NIR-A', 'AF-A']

for x in ChannelNames:

    tsample = sample.transform('hlog', channels=[x], b=500.0)

    figure();
    grid(True);
    tsample.plot([x], color = 'green', alpha = 0.7, bins = 100)

    tsample = sample.transform('hlog', channels=['FSC-H', 'SSC-H'], b=500.0)

    figure();
    grid(True);
    tsample.plot(['FSC-H', 'SSC-H'], color = 'green', alpha = 0.7, bins = 100)
```
